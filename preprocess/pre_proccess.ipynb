```python
# Loai bo trung lap
def dropDuplicate(data, subset):
    data = data.drop_duplicates(subset=subset)
```

```python
def standardRam(data, ramField):
    ls = []
    for ram in data[ramField]:
        ram = str(ram)
        ram = re.sub(r"[^0-9]", "",ram)
        ls.append(str(ram))
    data[ramField] = ls
    return data
```

```python
def standardPrice(data, priceField):
    ls = []
    for ram in zip(data[priceField]):
        ram = str(ram)
        ram = re.sub(r"[^0-9]", "",ram)
        print(ram)
        ls.append(str(ram))
    data[priceField] = ls
    return data
```
